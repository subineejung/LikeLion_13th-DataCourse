# 요일별 식사금액은 얼마나 될까?



```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
# 요일별 팁에 대해 알아보자
sns.boxplot(x = 'day', y = 'tip', data = tips)
plt.show()
```

```python
# 요일별 사이즈(식사인웢)에 대해 알아보자
sns.boxplot(x = 'day', y = 'size', data = tips)
plt.show()
```

```python
# 담배피는 사람, 안피는 사람과 팁의 관계
sns.boxplot(x = 'day', y ='tip', hue = 'smoker', data=tips)
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y='tip', hue='smoker', data=tips, size=7)
plt.show()
```

```python
fg = sns.load_dataset('flights')
fg
```

```python
sns.scatterplot(x = 'year', y = 'passengers', data=fg)
```

```python
sns.lmplot(x='year', y='passengers', hue='month', data=fg, size=7)
plt.show()
```

```python
fgpivot = fg.pivot('month', 'year', 'passengers')
fgpivot
```

```python
sns.heatmap(fgpivot, annot=True, fmt='d')
```
